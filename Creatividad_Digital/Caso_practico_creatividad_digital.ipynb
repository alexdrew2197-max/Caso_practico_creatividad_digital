## 3-Indicadores derivados (Seccion 3)  

```python
#Indicadores basicos:
df['ctr'] = df['clicks'] / df['impresiones']
df['cvr'] = df['conversiones'] / df['clicks']

#Indicadores de costo
df['cpc'] = df['costo_usd'] / df['clicks']
df['cpm'] = (df['costo_usd']/df['impresiones']) * 1000

#Indicador compuesto
df['roi'] = (df['ingresos_usd'] - df['costo_usd']) / df['costo_usd']
```

## 6- Estadistica descriptiva: dispersión (Seccion 5, continuacion)

```python
#Metricas a analizar
metricas = ['costo_usd','impresiones','clicks','conversiones','ingresos_usd','ctr','cvr','cpc','cpm','roi']
#Dataframe de dispersion, se tiene desviacion estandar, varianza, rango y rango IQR
dispersion = pd.DataFrame({
    'std': df[metricas].std(),
    'var': df[metricas].var(),
    'max - min': df[metricas].max() - df[metricas].min(),
    'Q3 - Q1': df[metricas].quantile(0.75) - df[metricas].quantile(0.25)
})

print(dispersion)
```

## 7- Deteccion de outliers con IQR (Seccion 6)

```python
#Definir Q1, Q3 e IQR
q1_ctr = df.ctr.quantile(0.25)
q3_ctr = df.ctr.quantile(0.75)
IQR_ctr = q3_ctr - q1_ctr

#Definir limites inferior y superior
Q1 = q1_ctr - 1.5 * IQR_ctr
Q3 = q3_ctr + 1.5 * IQR_ctr

#Guardar los outliers en outliers_ctr
outliers_ctr = df[(df.ctr < Q1) | (df.ctr > Q3)]
print(f"Outliers: {len(outliers_ctr)}\n")

print(outliers_ctr)
```

1. Para outliers de ctr solo hay 3.
2. Los valores que tienen los outliers varian mucho dependiendo de la categoria, por ejemplo, en ingresos unos son valores altos, mientras que los demas son bajos, pero en las otras categorias, estos tienen valores que podrian considerarse elevados. En palabras simples, los valores no tienen tanta correlacion.
3. Las plataformas en las que se encuetran todos los outliers es en TIKTOK

```python
#AHORA HACIENDO EL ANALISIS PARA ROI
#Definir Q1, Q3 e IQR
q1_roi = df.roi.quantile(0.25)
q3_roi = df.roi.quantile(0.75)
IQR_roi = q3_roi - q1_roi

#Definir limites inferior y superior
Q1 = q1_roi - 1.5 * IQR_roi
Q3 = q3_roi + 1.5 * IQR_roi

#Guardar los outliers en outliers_roi
outliers_roi = df[(df.roi < Q1) | (df.roi > Q3)]
print(f"Outliers: {len(outliers_roi)}\n")

print(outliers_roi)
```

## 8- Visualizacion (Seccion 7)

```python
#Definimos el tamaño de la grafica
plt.figure(figsize=(8, 5))

#plt.hist crea el histograma, usando .dropna evitamos usar valores que puedan ser nulos
plt.hist(df['roi'].dropna(), bins=50, color='skyblue', edgecolor='black')

#Agregamos titulo, label del eje X e Y
plt.title('Distribución del Retorno de Inversión (ROI)')
plt.xlabel('ROI')
plt.ylabel('Frecuencia (Cantidad de Campañas)')

#Hace el cuadriculado a la grafica  
plt.grid(axis='y', alpha=0.75)
#Enseña la grafica
plt.show()
```

### ¿Que muestra la grafica?
La grafica nos muestra que, un gran numero de campañas han resultado en un ROI demasiado bajo, solo siendo algunas de ellas las que han podido ser exitosas y regresar más del doble de lo invertido en ellas.

```python
#Definimos el tamaño de la grafica
plt.figure(figsize=(8, 5))

# Extrae las plataformas, sin repetir
plataformas = df['plataforma'].dropna().unique()

#Agrupa los datos de las plataformas a un dato CTR
datos_ctr = [df[df['plataforma'] == plat]['ctr'].dropna() for plat in plataformas]

#Crea el boxplot a partir de los datos_ctr y los nombres de las plataformas
plt.boxplot(datos_ctr, labels=plataformas)

#Añade el titulo y los labels al eje X e Y
plt.title('Dispersión del CTR por Plataforma')
plt.xlabel('\nPlataforma')
plt.ylabel('Click-Through Rate (CTR)')

#Enseña el grafico
plt.show()
```

### ¿Que muestra la grafica?
La grafica nos muestra que plataformas como instagram y tiktok tuvieron casos de outliers de bajo rendimiento, mientras que plataformas como Facebook tuvieron casos de outliers positivos.

```python
#Definimos el tamaño de la grafica
plt.figure(figsize=(8, 5))

#En la funcion ctr_promedio se agrupa el nombre del formato y el promedio ctr
ctr_promedio = df.groupby('formato')['ctr'].mean()

#Plt.bar ayuda a crear la tabla
#.index nos da los nombres (formato) y .values nos da los valores (ctr)
plt.bar(ctr_promedio.index, ctr_promedio.values, color='coral', edgecolor='black')

#Creamos el titulo y los labels de los ejes X e Y
plt.title('CTR Promedio según el Formato del Anuncio')
plt.xlabel('Formato')
plt.ylabel('CTR Promedio')

#Enseñamos la grafica
plt.show()
```

### ¿Que muestra la grafica?

La grafica nos muestra que, el contenido de reels y story, tienen un mejor CTR, comparado con otras formas de contenido. Esto significa que el contenido de reels y story es más relevante para las audiencias.

```python
#Definimos el tamaño de la grafica
plt.figure(figsize=(8, 5))

#Juntamos en una funcion los ingresos totales con las fechas de cada mes
ingresos_mensuales = df.groupby(df['fecha'].dt.to_period("M"))['ingresos_usd'].sum()

#plt.plot nos ayuda a crear la serie de tiempo
#Tomamos las fechas como strings por medio de .index y los ingresos totales con .values
plt.plot(ingresos_mensuales.index.astype(str), ingresos_mensuales.values, marker = 'o', label = 'ingresos totales (USD)')

#Agreage el titulo y los labels al eje X e Y
plt.title('Evolución de Ingresos Mensuales Totales')
plt.xlabel('Mes')
plt.ylabel('Ingresos Totales (USD)')

# Gira las fechas para que no se encimen
plt.xticks(rotation=45)

#Enseña lo que pusimos en el label
plt.legend() 

#Nos enseña la grafica
plt.show()
```

### ¿Que muestra la grafica?

La grafica nos muestra como en los dos primeros meses del 2025 se tuvieron unos ingresos mensuales positivos, sin embargo, en el 3 y 4 mes los ingresos tuvieron un pequeño declive, pero a pesar de eso, en los meses 5 y 6 se lograron recuperar los ingresos perdidos e incluso crear una ganancia mucho mayor que la de los primeros 2 meses.

## 9- Merge con tabla de clientes (Nueva sección)

```python
#Funcion para hacer un left merge
clientes = pd.read_csv('datos/clientes.csv')
m1 = pd.merge(df, clientes, on='cliente', how='left')

#Verificamos que el numero de filas sean iguales
print(f"Filas en lista original: {len(df)}")
print(f"Filas en lista m1: {len(m1)}")
```

```python
#Funcion para hacer un inner merge
m2 = pd.merge(df, clientes, on='cliente', how='inner')

#Verificamos el numero de filas en comparacion con m1
print(f"Filas en lista m1: {len(m1)}")
print(f"Filas en lista m2: {len(m2)}")
```

```python
#Funcion para hacer un outer merge
m3 = pd.merge(df, clientes, on='cliente', how='outer')

#Verificamos el numero de filas en comparacion con m1
print(f"Filas en lista m1: {len(m1)}")
print(f"Filas en lista m3: {len(m3)}")

#Clientes que no tienen campañas en el dataset principal
m3[m3['id_campana'].isna()]
```

## 10- Analisis post-merge

```python
#Agrupamos industria por el promedio de ROI
pm1 = m1.groupby('industria')['roi'].mean().sort_values(ascending = False)
print(pm1)
```

La industria que tiene mejor desempeño es la de Retail

```python
#Agrupamos el nivel de contracto junto el propedio de CTR
pm2 = m1.groupby('nivel_contrato')['ctr'].mean().sort_values(ascending=False)
print(pm2)
```

Los clientes de enterprise SI obtienen un mejor CTR

```python

'''
Hacemos la asignacion de campañas de clientes de tamaño grande
Usamos .agg para poder sumar cuantas campañas tienen un roi > 0 
Usamos mean para el porcentaje de exito
Usamos count para el numero de campañas totales
'''
pm3 = (m1[m1['tamano'] == 'Grande']['roi'] > 0).agg(['sum', 'mean', 'count'])

print(f"Campañas exitosas: {pm3['sum']}")
print(f"Porcentaje de éxito: {pm3['mean'] * 100:.2f}% (de un total de {pm3['count']} campañas)")
```
