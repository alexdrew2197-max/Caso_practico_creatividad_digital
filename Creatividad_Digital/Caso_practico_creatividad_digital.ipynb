## 1- Carga de datos (Seccion 1)

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
def cargar_datos(path):
 path_file = 'datos/campanas_creatividad_digital.csv'
 path_complete = os.path.join(os.getcwd(), path_file)
 df = pd.read_csv(path_complete, parse_dates = ['fecha'])
 return df
```

```python
path = 'datos/campanas_creatividad_digital.csv'
df = cargar_datos(path)
```
